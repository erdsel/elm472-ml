```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def load_txt_function(file_name):
    return np.loadtxt(file_name)
```

```python
def compute_parameters(x_training_data, r_data):
    N = len(x_training_data)
    
    # Ortalamaları hesapla
    x_mean_value = np.mean(x_training_data)
    r_mean_value = np.mean(r_data)
    xy_mean = np.mean(x_training_data * r_data)
    
    parameter_w1 = (np.sum(x_training_data * r_data) - x_mean_value * r_mean_value * N) / (np.sum(x_training_data**2) - N * x_mean_value**2)
    parameter_w0 = r_mean_value - parameter_w1 * x_mean_value
    
    return parameter_w1, parameter_w0

```

```python
def plot_data_and_model(x_training_data, r_data, parameter_w1, parameter_w0):
    # Grafik boyutunu ayarlama
    plt.figure(figsize=(5, 7))
    
    # Veriyi dağılım grafiği olarak çizme
    plt.scatter(x_training_data, r_data, color='purple', label='Eğitim Veri Noktaları')
    
    plt.plot(x_training_data, parameter_w1*x_training_data + parameter_w0, 'c', label=f'Model: y = {parameter_w1:.3f}x + {parameter_w0:.3f}')
    
    # Grafik başlığı ve eksen isimleri
    plt.title('Eğitim Veri Noktaları ve Lineer Regresyon Modeli')
    
    # Eksenleri veri aralığına sıkıca uyum sağlama
    plt.axis('tight')
    
    # Legend konumunu üst sağ köşeye taşı
    plt.legend(loc='lower right')
    
    # Grafiği göster
    plt.show()
```

```python
if __name__ == "__main__":
    
    data = load_txt_function('samples.txt')
    x_training_data, r_data = data[:, 0], data[:, 1]
    
    parameter_w1, parameter_w0 = compute_parameters(x_training_data, r_data)
    
    print(f"(w1): {parameter_w1:.3f}")
    print(f"(w0): {parameter_w0:.3f}")
    
    plot_data_and_model(x_training_data, r_data, parameter_w1, parameter_w0)
```
